```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
months = np.array(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'])
sales = np.array([150, 180, 220, 170, 200, 230])

# Unethical Visualization: Truncated y-axis
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(months, sales, marker='o')
plt.ylim(150, 240)  # Truncated y-axis
plt.title('Misleading Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

# Ethical Visualization: Full y-axis
plt.subplot(1, 2, 2)
plt.plot(months, sales, marker='o')
plt.ylim(0, 250)  # Full y-axis
plt.title('Accurate Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```
